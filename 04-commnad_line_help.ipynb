### コマンドラインのヘルプについて

#### 全体のヘルプ

```bash
nova help
```

#### サブコマンドのヘルプ

```bash
nova help boot
```

#### いくつかの例

```bash
neutron help
```

```bash
neutron help net-create
```

```bash
heat help
```

```bash
heat help stack-create
```

#### 例外 openstack コマンド

```bash
openstack help
```

```bash
openstack server -h
```

```bash
openstack server create -h
```
